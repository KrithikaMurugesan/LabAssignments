Exercise 2: Polynomial Regression

```python
#TASK A
#Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

#Create x and y
x = np.random.normal(1,0.05,100)
x = x.reshape(-1,1)

eps = np.random.randint(200,size = 100)
eps = eps.reshape(-1,1)
y = ((1.3*x*x)+(4.8*x)+8)+eps
```

```python
#Polynomial regression
def polyRegression(x,y,deg):
    #Generating polynomial features of given degree
    poly = PolynomialFeatures(degree=deg)
    xp = poly.fit_transform(x)

    # Instantiate
    lg = LinearRegression()
    
    # Fit
    lg.fit(xp, y)

    #Predict
    yHat = lg.predict(xp)
    
    #Sort x to make the plot more explanatory
    xPlot0 = np.sort(x,axis = 0)
    yPlot = ((1.3*x*x)+(4.8*xPlot0)+8)+eps
    xPlot = poly.fit_transform(xPlot0)
    #Using previous beta's so that sorting does not affect prediction
    yHatPlot = lg.predict(xPlot)
    #Original data plot
    plt.scatter(xPlot0,yPlot,color = "green")
    #Prediction line
    plt.plot(xPlot0,yHatPlot,color = "blue")
    plt.xlabel("X")
    plt.ylabel("Y")
    #plt.xlim(1.00,1.01)
    plt.title("Prediction curve")
    plt.show()
```

```python
deg = [1,2,7,10,16,100]
for each in deg:
    print("\nPolynomial degree : ",each)
    polyRegression(x,y,each)
```

```python
#Ridge regression
def ridgePoly(xp,y,alpha):
    #Setting degree to 10
    poly = PolynomialFeatures(degree=10)
    xp = poly.fit_transform(x)

    #RidgeRegression
    ridgeModel = Ridge(alpha=alpha)
    ridgeModel.fit(xp, y)
    yHat = ridgeModel.predict(xp)
    
    #Sort x to make the plot more explanatory
    xPlot0 = np.sort(x,axis = 0)
    yPlot = ((1.3*x*x)+(4.8*xPlot0)+8)+eps
    xPlot = poly.fit_transform(xPlot0)
    #Using previous beta's so that sorting does not affect prediction
    yHatPlot = ridgeModel.predict(xPlot)
    #Original data plot
    plt.scatter(xPlot0,yPlot,color = "green")
    #Prediction line
    plt.plot(xPlot0,yHatPlot,color = "blue")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Prediction curve")
    plt.show()
```

```python
alpha = [0,0.00000000001,0.000001,1]
for each in alpha:
    print("\nAlpha : ",each)
    ridgePoly(x,y,each)
```
